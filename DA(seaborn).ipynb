```python
import seaborn as s
```

```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
df=pd.read_csv('tips.csv')
df.sample(10)
```

```python
s.distplot(df['total_bill'])
```

```python
s.distplot(df['total_bill'],kde=False,bins=30)
```

```python
s.jointplot(x='total_bill',y='tip',data=df)
```

```python
s.jointplot(x='total_bill',y='tip',data=df,kind='hex',color='black')
```

```python
s.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
s.pairplot(df)
```

```python
s.pairplot(df,hue='sex',palette='coolwarm')
```

```python
s.rugplot(df['total_bill'])
```

```python
s.barplot(x='sex',y='total_bill',data=df)
```

```python
s.barplot(x='sex',y='total_bill',data=df,estimator=np.std)
```

```python
s.countplot(x='sex',data=df)
```

```python
s.boxplot(x='day',y='total_bill',data=df,palette='rainbow')#overfit
```

```python
s.violinplot(x='day',y='total_bill',data=df,palette='rainbow',hue='sex',split=True)
```

```python
s.stripplot(x='day',y='total_bill',data=df,jitter=False)
```

```python
s.stripplot(x='day',y='total_bill',data=df)
```

```python
s.stripplot(x='day',y='total_bill',data=df,jitter=True,hue='sex')
```

```python
s.stripplot(x='day',y='total_bill',data=df,jitter=True,hue='sex',split=True)#split changed to dodge
```

```python
s.swarmplot(x='day',y='total_bill',data=df,hue='sex',dodge=True)#split changed to dodge
```

```python
s.violinplot(x='tip',y='day',data=df,palette='rainbow')
s.swarmplot(x='tip',y='day',data=df,color='black',size=3)
```

```python
s.catplot(x='sex',y='total_bill',data=df,kind='bar') #factorplot changed to catplot
```
