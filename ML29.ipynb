# Random Module

The random module provides nice functions to generate random numbers

```python
#Random numbers between[0,1) of shape 2,2
import numpy as n
print(n.random.rand(3,3))
```

```python
#Normal distribution with mean=0 and varience=1 of shape 2,2
print(n.random.randn(2,2))
```

```python
#random integers b/w [0,10) of shape 2,2
print(n.random.randint(0,10, size=[2,2])) 
```

```python
#one random number b/w [0,1)
print(n.random.random())
```

```python
a=n.random.random(size=[3,3])
print(a)
```

```python
print(a.sum())
```
