```python
%matplotlib inline
import numpy as np
from pylab import *
import pandas as pd
```

```python
data = pd.read_csv("titanic.csv")
data.columns = ['Survived','Pclass','Name','Sex','Age','Siblings/Spouses Aboard','Parents/Children Aboard','Fare']
data.head()
```

```python
dt=data.drop(['Name'],1)
dt.head()
```

```python
gender=pd.get_dummies(dt['Sex'],drop_first=True)
```

```python
dt.drop(['Sex'],axis=1,inplace=True)
```

```python
dt=pd.concat([dt,gender],axis=1)
dt.head()
```

```python
dt.corr()
```

```python
X=dt.ix[:,(1,2,3,4,5,6)].values
y=dt.ix[:,0].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=25)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
```

```python
ld=LogisticRegression()
```

```python
ld.fit(X_train,y_train)
```

```python
ld.score(X_test,y_test)
```
